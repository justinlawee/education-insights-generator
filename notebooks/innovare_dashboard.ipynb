```
client = bigquery.Client(project='nw-ai-lab')
```

```
datasets = list(client.list_datasets())

if datasets:
    print("Datasets in project:")
    for dataset in datasets:
        print(f"- {dataset.dataset_id}")
else:
    print("No datasets found.")
```

```
# Replace 'your-project-id' if needed
project_id = client.project

# Dataset names
datasets = ['illinois_2024_nw', 'illinois_2025_nw']

for dataset_name in datasets:
    print(f"\nSchema for dataset: {dataset_name}")

    dataset_ref = bigquery.DatasetReference(project_id, dataset_name)
    tables = list(client.list_tables(dataset_ref))

    for table in tables:
        table_ref = dataset_ref.table(table.table_id)
        table_obj = client.get_table(table_ref)

        print(f"\nTable: {table.table_id}")
        for schema_field in table_obj.schema:
            print(f" - {schema_field.name} ({schema_field.field_type})")
```

```
query = """
    SELECT *
    FROM `illinois_2025_nw`.`iar_math`
"""
df = client.query(query).to_dataframe()
df.head()
```
